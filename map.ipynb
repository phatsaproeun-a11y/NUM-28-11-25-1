##Part 5: Getting Items

__getitem__ lets us use value = map[key] syntax

Step 1: Loop through all entries to find matching key

Step 2: If found, return the value

Step 3: If not found, raise KeyError (like Python's built-in dict)

Time Complexity: O(n) - linear search

```python
def __getitem__(self,key:k ) -> V:
  """Get value for key -o(n)"""
  for entry in self._entries:
    if entry.key == key:
      return entry.value
  raise KeyError(f"key {key} not found")
```

##Part 6: Deleting Items
__delitem__ lets us use del map[key] syntax

enumerate() gives us both index (i) and value (entry)

Step 1: Find the entry with matching key

Step 2: Use the index to delete it from the list

Step 3: If not found, raise error

```python
def __delitem__(self,key : k) -> None:
  """Remove key_value pair o(n)"""
  for i, entry in enumerate(self._entries):
    if entry.key == key:
      del self._entries[i]
      return
  raise KeyError(f"key {key} not found")
```

##Part 7: Checking if Key Exists
__contains__ lets us use if key in map syntax

Returns True if key exists, False if not

```python
def __contains__(self, key: k) -> bool:
  """Check if key exists -o(n)"""
  for entry in  self._entries :
    if entry.key == key:
      return True
  return False
```
